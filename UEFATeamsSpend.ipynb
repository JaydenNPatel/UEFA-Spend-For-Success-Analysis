```python
import pandas as pd
# First I imported pandas so I could manipulate and view the dataset

top_50_info = pd.read_csv("Top50EuropeSoccerSpending.csv")
# I then read in my csv file

top_50_info
# Now I will start a cleaning process for this particular set
```

I see that I did not specify what unit or proportion my 'balance' column is, so I will fix that in the next cell

```python
top_50_info = top_50_info.rename(columns={'Balance': 'Balance (€) in Millions'})
top_50_info.head()
# I renamed the column, so it is more descriptive
```

```python
# I will now check if all of the datatypes are correct
top_50_info.info()
```

```python
# Expenditures seems to be an object because I accidentally put
# a € in the expenditures for 'Bayer Leverkusen'. I will fix that now
top_50_info['Expenditures (€) in Millions'] = top_50_info['Expenditures (€) in Millions'].replace(['€ 199.65'],['199.65'])
top_50_info.info()
```

```python
top_50_info['Expenditures (€) in Millions'] = top_50_info['Expenditures (€) in Millions'].astype(float)
top_50_info.info()
```

Now the numbers are all set to the right data types (a float)

```python
# Going through the 'Balance (€) in Millions' column, the team "Bayer Leverkusen"'s balance seems to be 0. Since I used an Excel 
# shortcut to subtract expenditures by income, I may have inputted the same value for income 
# that I did for expenditures or vice versa.

# After checking, I inputted the same expenditure into the income column. I will change it below
top_50_info['Income (€) in Millions'] = top_50_info['Income (€) in Millions'].replace(436.62, 573.67)
top_50_info['Balance (€) in Millions'] = top_50_info['Balance (€) in Millions'].replace(0.00, -137.05)
```

```python
# Now there are no mistakenly inputted values in the dataset
top_50_info
```
